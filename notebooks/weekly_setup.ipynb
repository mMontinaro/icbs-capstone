```python
import nbformat

def append_weekly_cells(week_number: int):
    """
    Appends 5 cells at the end of each function notebook
    1. Type: Markdown   ; Content: Week <week_number>
    2. Type: Code       ; Content: x, y = hd.load_and_append(<function_id>, <week_number>)
    3. Type: Code       ; Content: dd.create_and_display_df(x, y, cols)
    4. Type: Code       ; Content: <empty> (tune params and execute predictions)
    5. Type: Markdown   ; Content: <empty> (notes and observations)
    """

    for function_id in range(1,9):
        path = f"function{function_id}.ipynb"

        with open(path, 'r', encoding='utf-8') as f:
            nb = nbformat.read(f, as_version=4)

        already_exists = any(
            cell.cell_type == "markdown" and f"# Week {WEEK}" in cell.source
            for cell in nb.cells
        )
        if already_exists:
            print(f"[Weekly setup] --- Week {WEEK} already exists in {path}")
            continue

        new_cells = [
            nbformat.v4.new_markdown_cell(f"## Week {WEEK}"),
            nbformat.v4.new_code_cell(f"x, y = hd.load_and_append({function_id}, {WEEK})"),
            nbformat.v4.new_code_cell(f"dd.create_and_display_df(x, y, cols)"),
            nbformat.v4.new_code_cell(""),
            nbformat.v4.new_markdown_cell("")
        ]

        nb.cells.extend(new_cells)

        with open(path, "w", encoding="utf-8") as f:
            nbformat.write(nb, f)

        print(f"[Weekly setup] --- Appended Week {WEEK} cells to function{function_id}.ipynb")

append_weekly_cells(WEEK)
```
