```python
def lower_bound_15_percent(x):
    """
    Returns a number 15% smaller than the input x.
    """
    return x * (1 - 0.15)

acquisition = "ei"
beta = 2.5
xi = 0.005
noise = 1e-8
length_scale=0.005
n_candidates=20000
length_scale_bounds=(1e-2, 1)
lower_bound = x[:,0].min()
print("Lowest x1:", lower_bound)
lower_bound = lower_bound_15_percent(lower_bound)

bo = op.BayesianOptimizer(
    x, 
    y, 
    acquisition=acquisition, 
    beta=beta, 
    noise=noise,
    length_scale=length_scale,
    xi=xi,
    length_scale_bounds=length_scale_bounds
)
next = bo.propose(n_candidates, lower_bound=lower_bound)
op.print_next_submission(next)
```

* switched to ei, want to capitalize on the known optimum
* had to tweek the lower bound. x1 kept giving me 0, so i decided to fix it to the lowest x1 recorded +- 15%
